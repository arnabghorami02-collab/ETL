# DATA EXTRACTION IN ETL ASSIGNMENT
Question 1 : Describe different types of data sources used in ETL with suitable examples.

Sol :    
     ETL uses multiple data sources:

- Relational Databases – Structured data in tables.
Example: MySQL, Oracle, SQL Server.

- Flat Files – Text-based files.
Example: CSV, TXT.

- Spreadsheets – User-managed data files.
Example: Excel (.xlsx).

- APIs / Web Services – Real-time external data.
Example: Weather API, Payment Gateway API.

- Cloud Storage / Data Lakes – Large-scale raw data storage.
Example: AWS S3, Azure Blob.

- NoSQL Databases – Semi-structured/unstructured data.
Example: MongoDB, Cassandra

Question 2 : What is data extraction? Explain its role in the ETL pipeline.

Sol :  Data extraction is the process of collecting data from different source systems for further processing.

Role in ETL:

- First step of ETL.
- Retrieves raw data from sources.
- Ensures required data is available for transformation.
- Can be full extraction or incremental extraction.

 Question 4 : Explain the steps involved in extracting data from a relational database.
 Sol :    

- Identify source database.
- Connect using credentials.
- Select required tables/views.
- Write SQL query.
- Extract full or incremental data.
- Validate extracted records.
- Store in staging area.
- Log extraction status/errors.

Question 5 : Explain three common challenges faced during data extraction.

Sol :        

- Poor Data Quality – Missing or duplicate records.
- Performance Issues – Large datasets take time.
- Schema Changes – Source table structure changes unexpectedly.

Question 6 : What are APIs? Explain how APIs help in real-time data extraction.

Sol :    

API (Application Programming Interface) allows systems to communicate and exchange data.


How APIs help:

- Fetch live data instantly.
- Connect cloud apps easily.
- Provide JSON/XML structured data.
- Support automation and scheduled pulls.

Example: Fetching live stock prices using finance API.

Question 7 : Why are databases preferred for enterprise-level data extraction?

Sol :   

- Handle very large data volumes.
- Better security and access control.
- Fast querying using indexes.
- Reliable backup and recovery.
- Supports concurrent users.
- Easy integration with ETL tools.

Question 8 : What steps should an ETL developer take when extracting data from large CSV files (1GB+)?

Sol :        

- Use chunk/batch processing.
- Validate delimiter and encoding.
- Remove corrupt rows.
- Use parallel processing if possible.
- Load into staging table first.
- Monitor memory usage.
- Compress/archive old files.
- Maintain logs for errors and row counts.
